```python
# Remove leading/trailing spaces from column names
df.columns = df.columns.str.strip()

# Verify cleaned column names
print(df.columns)
```

```python
import folium
from folium.plugins import HeatMap

# Ensure Latitude and Longitude columns exist
if "latitude" in df.columns and "longitude" in df.columns:
    # Drop missing values
    df = df.dropna(subset=["latitude", "longitude"])

    # Create a Base Map (Centering around the dataset's mean lat/lon)
    map_center = [df["latitude"].mean(), df["longitude"].mean()]
    map_thermal = folium.Map(location=map_center, zoom_start=6)

    # Add HeatMap Layer
    heat_data = df[["latitude", "longitude"]].values.tolist()
    HeatMap(heat_data, radius=10, blur=15, max_zoom=1).add_to(map_thermal)

    # Save & Display Map
    map_thermal.save("thermal_map.html")
    print("Thermal Map generated! Open 'thermal_map.html' to view it.")

else:
    print("Latitude and Longitude columns not found in dataset!")
```

```python
import matplotlib.pyplot as plt

if "date" in df.columns and "new_confirmed" in df.columns:
    df["date"] = pd.to_datetime(df["date"])
    daily_cases = df.groupby("date")["new_confirmed"].sum().reset_index()

    plt.figure(figsize=(10, 5))
    plt.plot(daily_cases["date"], daily_cases["new_confirmed"], color="blue")
    plt.title("Daily New COVID-19 Cases")
    plt.xlabel("Date")
    plt.ylabel("New Cases")
    plt.xticks(rotation=45)
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.show()
else:
    print("Necessary columns not found!")
```

```python
if "country_name" in df.columns and "new_confirmed" in df.columns:
    top_countries = df.groupby("country_name")["new_confirmed"].sum().nlargest(10)

    plt.figure(figsize=(10, 5))
    plt.barh(top_countries.index, top_countries.values, color="green")
    plt.title("Top 10 Countries with Most COVID-19 Cases")
    plt.xlabel("Total Cases")
    plt.ylabel("Country")
    plt.show()
else:
    print("Necessary columns not found!")
```

```python
if "date" in df.columns and "new_confirmed" in df.columns:
    df["date"] = pd.to_datetime(df["date"])
    df["month"] = df["date"].dt.strftime("%Y-%m")  # Extract year-month
    monthly_cases = df.groupby("month")["new_confirmed"].sum()

    plt.figure(figsize=(12, 5))
    plt.bar(monthly_cases.index, monthly_cases.values, color="orange")
    plt.title("Monthly COVID-19 Cases")
    plt.xlabel("Month")
    plt.ylabel("Total Cases")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Necessary columns not found!")
```
